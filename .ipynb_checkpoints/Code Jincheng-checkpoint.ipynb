# Part0. Problem Discription

We are given 26 Canadian Government bonds with different Maturity Date, the earlist is March 2020. We have to select 10 bonds out of them to construct yield curve, spot curve and forward curve for 0-5 years. The 10 bonds should have consecutive maturity date and they should be 0.5 years apart. However, two of the bonds are missing( The one who meatures on September 2022 and September 2023), so I have to use some other bonds with close maturity date and use interpolation to calculate those curves. This cause extra works and may make the code become quite messy. 

# Part1. Preparation

```python
#data10 = pd.read_csv('bond_price.csv', index_col = 0)
data10 =  pd.read_csv('bond_price2.csv', index_col = 0)
time = list(range(0,11))
for i in range(len(time)):
    time[i] = time[i]/2
time.pop(0)
time[5] = 2.75
time[7] = 3.75
```

### Create classes of bonds 
The reason that I used Object-Oriented Programming for this project is because it is easier to keep track of lots of information for a single bond, such as isin number, coupon payment, maturity date, etc. Some other ways of storing data can also be used here, such as using pandas DataFrame. 

Please note that the reason that I used two different classes is because we don't have enough types of bonds. To calculate the yield curve for 0-5 years using 10 bonds, we have to use 10 bonds with consecutive maturity dates and they should be 0.5 year apart. Note that the start date is Septermber 2019. However, two of the bonds are missing, so I have to use some other bonds with close maturity date and use interpolation to calculate those curves.

```python
# I create a class with all information of the bond
class bond_regular():
    def __init__(self, isin):
        self.isin = isin
        self.coupon = data10.loc[isin,:]['coupon']
        self.mature = pd.to_datetime(data10.loc[isin,:]['Mature'])
        # Stores clean(quoted) price of the bond at different times
        dict_price = {}
        for item in list(data10.columns)[3:]:
            dict_price[pd.to_datetime(item)] = data10.loc[isin,:][item]
        self.price = dict_price
        # Calculate accrued interest of the bond at different times
        dict_ai = {}
        for item in list(data10.columns)[3:]:
            dict_ai[pd.to_datetime(item)] = self.coupon * 100* ((pd.to_datetime(item) - pd.to_datetime("9/1/2019"))/np.timedelta64(1, 'D'))/365
        self.ai = dict_ai
        # Calculate dirty price of the bond at different times
        dict_dirty = {}
        for item in list(data10.columns)[3:]:
            # dirty price = quoted price + accrued interest
            dict_dirty[pd.to_datetime(item)] = dict_price[pd.to_datetime(item)] + dict_ai[pd.to_datetime(item)]
        self.dirty = dict_dirty
        # Find the number of coupon including principal payments
        self.coupon_number = round(((pd.to_datetime(self.mature)-pd.to_datetime("9/1/2019"))/np.timedelta64(1, 'M'))/6)
        # Find dates of coupon payment for the bond
        time_coupon = []
        total_lst = ["3/1/2020","9/1/2020","3/1/2021","9/1/2021","3/1/2022","9/1/2022","3/1/2023","9/1/2023","3/1/2024","9/1/2024","3/1/2025"]
        for item in total_lst:
            if pd.to_datetime(item) <= self.mature:
                time_coupon.append(item)
        self.coupon_date = time_coupon
    
```

```python
# I create a class with all information of the bond which has a irregular maturity date(namely,  6th  and 8th bond)
class bond_irr():
    def __init__(self, isin):
        self.isin = isin
        self.coupon = data10.loc[isin,:]['coupon']
        self.mature = pd.to_datetime(data10.loc[isin,:]['Mature'])
        # Stores clean(quoted) price of the bond at different times
        dict_price = {}
        for item in list(data10.columns)[3:]:
            dict_price[pd.to_datetime(item)] = data10.loc[isin,:][item]
        self.price = dict_price
        # Calculate accrued interest of the bond at different times
        dict_ai = {}
        for item in list(data10.columns)[3:]:
            # Please note that the two special bonds have coupon payments on Dec and June, while normal bond pays coupon on Sep and March
            dict_ai[pd.to_datetime(item)] = self.coupon * 100* ((pd.to_datetime(item) - pd.to_datetime("12/1/2019"))/np.timedelta64(1, 'D'))/365
        self.ai = dict_ai
        # Calculate dirty price of the bond at different times
        dict_dirty = {}
        for item in list(data10.columns)[3:]:
            # dirty price = quoted price + accrued interest
            dict_dirty[pd.to_datetime(item)] = dict_price[pd.to_datetime(item)] + dict_ai[pd.to_datetime(item)]
        self.dirty = dict_dirty
        # Find the number of coupon including principal payments
        self.coupon_number = round(((pd.to_datetime(self.mature)-pd.to_datetime("12/1/2019"))/np.timedelta64(1, 'M'))/6)
        # Find dates of coupon payment for the bond
        time_coupon = []
        total_lst = ["6/1/2020","12/1/2020","6/1/2021","12/1/2021","6/1/2022","12/1/2022","6/1/2023","12/1/2023","6/1/2024","12/1/2024"]
        for item in total_lst:
            if pd.to_datetime(item) <= self.mature:
                time_coupon.append(item)
        self.coupon_date = time_coupon
    
```

# Part2. Yield Curve Calculation
I calculated yield using $P_{dirty} = \sum_{i=1}^{m}C_{i}e^{-rt_{i}} $ to solve for r where $C_{i}$ denote cash flow at time $t_{i}$.

### Develop a function which we will find zeros later

```python
# Follow the formula above
def PV(r:float,bond,date)->float:
    price = 0
    time = []
    for i in bond.coupon_date:
        y = (pd.to_datetime(i) - date)/np.timedelta64(1, 'Y')
        time.append(y)
    for i in range(len(time)-1):
        price += (bond.coupon*100/2)*math.exp(time[i]*-1*r)
    price += (100+(bond.coupon*100/2))*math.exp((time[-1]*-1*r))
    return price - bond.dirty[date]
```

### Create a DataFrame to include yields for all bonds at different time

```python
a1 = np.zeros(shape=(11,10))
columns = list(data10.columns)[3:]
isin_lst1= list(data10.index)[:11]
isin_lst= list(data10.index)[:11]
df_yield1 = pd.DataFrame(a1,columns=columns, index = isin_lst1)
# For each bond at each time, we can find a value
for i in range(len(isin_lst1)):
    for j in range(len(columns)):
        if isin_lst1[i] in ["CA135087UM44","CA135087UT96"]:
            bond1 = bond_irr(isin_lst1[i])
        else:
            bond1 = bond_regular(isin_lst1[i])
        date1 = pd.to_datetime(columns[j])
        # Use newton method to find the zero
        df_yield1.iloc[i,j] = optimize.newton(PV, 0.01, args=(bond1,date1, ))
```

```python
# Create a list which stores time 
bond_last = bond_regular("CA135087K528")
time_ytm = []
for i in bond_last.coupon_date:
        y = (pd.to_datetime(i) - pd.to_datetime("1/2/2020"))/np.timedelta64(1, 'Y')
        time_ytm.append(y)
time_spot = time_ytm[:]
```

# Part3. Spot Curve Calculation
$t_{i}$ denote the time where i = 1,2,..,m, where m is the time of maturity for this bond, C denote each semiannual coupon payment. $r_{i}$ denotes spot rate at $t_{i}$.
        \begin{align}
        P_{\text{dirty}} = \sum_{i=1}^{m}Ce^{-r_{i}t_{i}} + 100*e^{-r_{m}t_{m}}\nonumber
        \end{align}
        What I do is that, I first get those $t_{i}$with i small, and then move forward to big i. Thus, given $t_{i}$ for i = 1,2,..,m-1 and P, we find $r_{m}$ as 
        \begin{align}
        r_{m} =\frac{-ln(\frac{P_{\text{dirty}}- \sum_{i=1}^{m-1}Ce^{-r_{i}t_{i}}}{100+C})}{t_{m}}\nonumber
        \end{align}

Please also note that since the sixth and eighth bonds are missing, we have to interpolate 

### Start from the first 5 bonds

```python
# Finding out spot rate for the first 5 
#recall time was defined before
time_plot = list(range(0,11))
for i in range(len(time_plot)):
    time_plot[i] = time_plot[i]/2
time_plot.pop(0)
r = np.zeros(shape=(11,10))
df_spot = pd.DataFrame(r,columns=columns, index = time_spot)
for j in range(len(columns)):
    # For each date
    r_list = []
    p_list = []
    # create the list of price for this day
    for b in range(len(isin_lst)):
        if isin_lst[b] in ["CA135087UM44","CA135087UT96"]:
            bond1 = bond_irr(isin_lst[b])
            p_list.append(bond1.dirty[pd.to_datetime(columns[j])])
        else:
            bond1 = bond_regular(isin_lst[b])
            p_list.append(bond1.dirty[pd.to_datetime(columns[j])])
    # create the time to discount
    time1 = []
    for i in bond1.coupon_date:
        y = (pd.to_datetime(i) - pd.to_datetime(columns[j]))/np.timedelta64(1, 'Y')
        time1.append(y)
    # We know the data for first 5 bonds are perfect, So I first find out spot rate for first period
    for l in range(5):
        bond_c = bond_regular(isin_lst[l])
        c = bond_c.coupon *100/2
        sum = 0
        for m in range(len(r_list)):
            sum += c * np.exp(-1*r_list[m]*time1[m])
        rate = -1*np.log((p_list[l]-sum)/(100+c))/time1[l]
        r_list.append(rate)
    for a in range(len(r_list)):
        df_spot.iloc[a,j] = r_list[a]
```

### Extrapolate to get the 6th, then 7th, extrapolate to get 8th, then the rest

```python
# In this code, I start to extrapolate the 6th spot rate, then use it to find next several spot rates
r_six = []
bond_six = bond_irr(isin_lst[5])
c = bond_six.coupon *100/2
for j in range(len(columns)):
    # For each date
    time2 = []
    for i in bond_six.coupon_date:
        y = (pd.to_datetime(i) - pd.to_datetime(columns[j]))/np.timedelta64(1, 'Y')
        time2.append(y)
    r_list = []
    for i in range(4):
        lst_inter = list(df_spot.loc[:,columns[j]])
        r_list.append(math.sqrt(lst_inter[i]*lst_inter[i+1])) # I choose geometric average for interpolation since dates are quite different
    sum = 0
    for m in range(len(r_list)):
            sum += c * np.exp(-1*r_list[m]*time2[m])
    p_six = bond_six.dirty[pd.to_datetime(columns[j])]
    rate = -1*np.log((p_six-sum)/(100+c))/time2[-1]
    r_six.append(rate)
# Thus, in r_six, I get the value for rate at 2022.6 and we need rate for 2022.9
# I use extrapolation since we don't have any information about the future rate or anything
# I use (2*2022.6rate + 1*2022.3 rate)/3
for index in range(10):
    df_spot.iloc[5,index] = (2*r_six[index]+df_spot.iloc[4,index])/3
```

```python
# Now, we have one more correct date bond
bond_seven = bond_regular(isin_lst[6])
c_seven = bond_seven.coupon *100/2
for j in range(len(columns)):
    # For each date
    time3 = []
    for i in bond_seven.coupon_date:
        y = (pd.to_datetime(i) - pd.to_datetime(columns[j]))/np.timedelta64(1, 'Y')
        time3.append(y)
    r_list = []
    p_list = []
    # create the list of price for this day
    # create the time to discount
    # we use past data to bootstrap for the current one 
    sum1 = 0
    for m in range(6):
            sum1 += c_seven * np.exp(-1*df_spot.iloc[m,j]*time3[m])
    p_seven = bond_seven.dirty[pd.to_datetime(columns[j])]
    rate = -1*np.log((p_seven-sum1)/(100+c_seven))/time3[-1]
    df_spot.iloc[6,j] = rate
```

```python
# Again, we have one incorrect date bond
r_eight = []
bond_eight = bond_irr(isin_lst[7])
c_eight = bond_eight.coupon *100/2
for j in range(len(columns)):
    # For each date
    time4 = []
    for i in bond_eight.coupon_date:
        y = (pd.to_datetime(i) - pd.to_datetime(columns[j]))/np.timedelta64(1, 'Y')
        time4.append(y)
    r_list = []
    for i in range(6):
        lst_inter = list(df_spot.loc[:,columns[j]])
        r_list.append(math.sqrt(lst_inter[i]*lst_inter[i+1])) # I choose geometric average for interpolation
    sum_eight = 0
    for m in range(len(r_list)):
            sum_eight += c_eight * np.exp(-1*r_list[m]*time4[m])
    p_eight = bond_eight.dirty[pd.to_datetime(columns[j])]
    rate = -1*np.log((p_eight-sum_eight)/(100+c_eight))/time4[-1]
    r_eight.append(rate)
# Thus, in r_six, I get the value for rate at 2022.6 and we need rate for 2022.9
# I use extrapolation since we don't have any information about the future rate or anything
# I use (2*2023.6rate + 1*2023.3 rate)/3
for index in range(10):
    df_spot.iloc[7,index] = (2*r_eight[index]+df_spot.iloc[6,index])/3
```

```python
# For the last three bonds
for j in range(len(columns)):
    # For each date
    r_list = []
    p_list = []
    # create the list of price for this day
    # create the time to discount
    for b in [8,9,10]:
        bond1 = bond_regular(isin_lst1[b])
        p_list.append(bond1.dirty[pd.to_datetime(columns[j])])
    time5 = []
    for i in bond1.coupon_date:
        y = (pd.to_datetime(i) - pd.to_datetime(columns[j]))/np.timedelta64(1, 'Y')
        time5.append(y)
    # We now know all the spot rate before
    for l in [8,9,10]:
        bond_c = bond_regular(isin_lst1[l])
        c = bond_c.coupon *100/2
        sum5 = 0
        for m in range(l):
            sum5 += c * np.exp(-1*df_spot.iloc[m,j]*time5[m])
        rate = -1*np.log((p_list[l-8]-sum5)/(100+c))/time5[l]
        r_list.append(rate)
    for a in [8,9,10]:
        df_spot.iloc[a,j] = r_list[a-8]
```

### Plot Spot Curve

```python
# Here, I will plot the data
fig = plt.figure()
for i in range(1,11):
    plt.plot(time_spot,df_spot.iloc[:,i-1],marker = "*")
#plt.legend(["Day " + str(i) for i in range(1, 11)])
plt.legend(columns)
plt.xlabel("Time")
plt.ylabel("Spot rate")
plt.title('Spot Curve')
fig.savefig('1.png')
```

# Part4. Forward Curve Calculation

Assuming continuously compounding, The formula I used was:
        \begin{align}
        f_{i,j} =  \frac{r_{j}t_{j}-r_{i}t_{i}}{t_{j}-t{i}}\nonumber
        \end{align}
        where $r_{j}$ denote spot rate at time j and $f_{i,j}$ denotes iyr-(j-i)yr forward rate from year i

```python
# I first interpolate to get spot rate for January each year
r1 = np.zeros(shape = (5,10))
time2 = list(range(1,6))
df_spot1 = pd.DataFrame(r1, columns = columns,index = time2)
for j in range(10):
    for i in range(5):
        df_spot1.iloc[i,j] =df_spot.iloc[i*2+1,j] * 1/3 + df_spot.iloc[i*2+2,j] * 2/3 
```

### Calculate forward rate

```python
# Calculating forward rate
r2= np.zeros(shape=(4,10))
time3 = list(range(1,5))
df_forward = pd.DataFrame(r2,columns=columns, index = time3 )
for i in range(10):
    # for each day
    for j in range(4):
        f_rate = ((j+2)*df_spot1.iloc[j+1,i]-df_spot1.iloc[0,i])/(j+1)
        df_forward.iloc[j,i] = f_rate
```

### Plot Foward Curve

```python
# plot forward rate
fig3 = plt.figure()
x_forward = ["1yr-1yr","1yr-2yr","1yr-3yr","1yr-4yr"]
for i in range(1,11):
    plt.plot(x_forward,df_forward.iloc[:,i-1],marker = "*")
#plt.legend(["Day " + str(i) for i in range(1, 11)])
plt.legend(columns)
plt.xlabel("1yr-(value)yr")
plt.ylabel("Forward rate")
plt.title('Forward Curve')
fig3.savefig('3.png')
```
